```python
import pandas as pd

data = {
    'Name': ['Sohail','Adeep','Abhi','Pradeep','Ashok','Ravi','Sneha','Kiran','Meena','Arjun'],
    'Marks': [87,98,45,76,39,92,65,55,81,73],
    'Branch': ['CSE','CSM','CAI','CSE','CSM','CAI','CSE','CSM','CAI','CSE'],
    'City': ['Guntur','Hyd','Vizag','Guntur','Hyd','Vizag','Guntur','Hyd','Vizag','Guntur'],
    'Gender': ['M','M','M','M','M','M','F','F','F','M']
}

df = pd.DataFrame(data)
print(df)
```

P1. Add Result column
Marks ≥ 50 → Pass
else → Fail

```python
df['Result']=df['Marks'].apply(lambda x: "Pass" if x>=50 else "Fail")
print(df)
```

 P2. Add Grade column
≥85 → A
≥70 → B
else → C

```python
def grade(x):
    if x>=85:
        return 'A'
    elif x>=70:
        return 'B'
    else:
        return 'C'
df['Grade']=df['Marks'].apply(grade)
print(df)
```

P3. Find average marks per Branch

👉 (groupby)

```python
avg_branch=df.groupby('Branch')['Marks'].mean().reset_index()
print(avg_branch)
```

P4. Find total students per City

```python
df['City'].value_counts()
```

P5. Find highest marks in each Branch

```python
high_each_brn=df.groupby('Branch')['Marks'].max()
print(high_each_brn)
```

P6. Pivot Table

Create table:

Rows → Branch
Columns → City
Values → Average Marks

```python
brn_c_avg=df.pivot_table(
    index='Branch',
    columns='City',
    values='Marks',
    aggfunc='mean',
    fill_value=0
)
print(brn_c_avg)
```

P7. Top Performers

Show students with Marks > 80
Name, Marks, Branch

```python
t_per=df[df['Marks']>80]
print(t_per[['Name','Marks','Branch']])
```

P8. Multi-condition filter

Students:

From CSE branch
AND Marks > 70

```python
cse_n_t=df[(df['Branch']=='CSE') & (df['Marks']>70)]
print(cse_n_t)
```

P9. Find which Branch performs best

👉 (based on average marks)

```python
best_brn=df.groupby('Branch')['Marks'].mean()
print(best_brn)
```

P10. Gender Analysis
Average marks of Male vs Female

👉 (this is real-world business question)

```python
ave_gender=df.groupby('Gender')['Marks'].mean()
print(ave_gender)
```

P11. Sort students by Marks (descending)

```python
print(df.sort_values(by='Marks',ascending=False))
```

P12. Export result to CSV

```python
df.to_csv('student_analysis.csv', index=False)
```
